# AI News Finder Daily Run

This notebook clones the repository, installs the dependencies it needs, and runs the daily AI news pipeline on Kaggle.

```python
import os

os.environ["AI_NEWS_USE_HF"] = "1"
os.environ["AI_NEWS_USE_SUMMARY_MODEL"] = "1"
os.environ["AI_NEWS_SUMMARY_MODEL"] = "Qwen/Qwen2.5-1.5B-Instruct"
os.environ["AI_NEWS_ENRICH_STORIES"] = "1"
```

```python
import pathlib
import shutil
import subprocess
import sys

repo_dir = pathlib.Path('/kaggle/working/news-update')
if repo_dir.exists():
    shutil.rmtree(repo_dir)

subprocess.run(['git', 'clone', '--depth', '1', 'https://github.com/aisangh/news-update.git', str(repo_dir)], check=True)
subprocess.run([
    sys.executable,
    '-m',
    'pip',
    'install',
    '-q',
    '-r',
    str(repo_dir / 'requirements.txt'),
    'sentence-transformers',
    'transformers',
    'accelerate',
    'sentencepiece',
], check=True)
subprocess.run([
    sys.executable,
    str(repo_dir / 'kaggle' / 'launch.py'),
    '--days',
    '2',
    '--reports-dir',
    '/kaggle/working/reports',
], check=True, cwd=str(repo_dir))
```
